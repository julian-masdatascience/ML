# .plot()

Múltiples parámetros

```python
plt.plot(x, y, style, label, linewidth=1, alpha=1)
```

- 'style': cadena de texto con formato (color, tipo...) \[color\]\[marker\]\[line\]
- 'label': referencia para leyenda
- 'linewidth': brocha
- 'alpha': transparencia

Otros: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html


````
y_data = np.random.randint(100,size=(10))
x_data = np.arange(10)
fig,axis = plt.subplots(1,1)
axis.plot(x_data,y_data,'go-',label='green',linewidth=5,alpha=0.2) # creación del gráfico
axis.legend(loc='best') # representar la leyenda
plt.show()
````

### Representar más de un gráfico en una subfigura

````
y_data = np.random.randint(100,size=(10))
x_data = np.arange(10)
fig,axis = plt.subplots(1,1)
axis.plot(x_data,np.random.randint(100,size=(10)),'go-',label='green',linewidth=1,alpha=0.2) # creación del gráfico
axis.plot(x_data,np.random.randint(100,size=(10)),'bx-.',label='blue',linewidth=1) # creación del gráfico
axis.legend(loc='best')
plt.show() 
````

# Anotaciones

Formas de añadir información sobre el gráfico

```python
axes.text(x,y,’texto’)
axes.annotate(texto, xy, xytext, xycoords, arrowprops)
```

[Más métodos](https://matplotlib.org/users/annotations.html)

````
y_data = np.random.randint(100,size=(10))
x_data = np.arange(10)
fig,axis = plt.subplots(1,1)
axis.plot(x_data,y_data)
axis.text(1,60,'Bajo Ingreso') # coordenadas igual a los datos
axis.annotate('outlier - Anomalia', xy=(3, 70), xytext=(2, 25), arrowprops = dict(facecolor='black', shrink=0.01)) # anotación completa
plt.show() 
````

# Tipos de gráficos

Ejemplos disponibles como referencia
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html

````
# Horizontal bar
fig, ax = plt.subplots()
# Example data
people = ('PhD', 'Master', 'Profesional', 'N/A')
y_pos = np.arange(len(people))
percentage = [5,45,25,30]

ax.barh(y_pos, percentage, align='center',
        color='blue', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
# ax.invert_yaxis()  # Invierte orden del eje Y
ax.set_xlabel('% Fuerza de trabajo')
ax.set_title('Nivel de educación de la compañia')
plt.show()
````

````
# stack plot
eje_x = [2008, 2009, 2010, 2011, 2012]
iOS = [35, 25, 20, 17, 18]
Android = [0, 15, 35, 45, 60]
WindowsPhone = [1, 3, 5, 5, 2]
Others = [64,57,40,33,20]

y = np.vstack([iOS, Android, WindowsPhone,Others])

labels = ["iOS ", "Android", "Windows Phone","Otros"]
fig, ax = plt.subplots()
ax.stackplot(eje_x, iOS, Android, WindowsPhone, Others,labels=labels)
ax.legend(loc='best')
#ax.legend(loc='upper left')
plt.show()
````

````
# Scatter plot
from numpy.random import rand
fig, ax = plt.subplots()
for color in ['red', 'green', 'blue']:
    n = 750
    x, y = rand(2, n)
    scale = 25.0 * rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color,
               alpha=0.3, edgecolors='none')

ax.legend()
plt.show()
````

````
import pandas as pd
rand_matrix = np.random.randint(100,size=(10,3))
frame = pd.DataFrame(rand_matrix , columns=list('ABC'))
frame
````

````
frame.plot()
plt.show() # aún es necesario llamar a show() para mostrar los gráficos

# Qué parte del DataFrame se utiliza para:
#   Datos en el eje X
#   Datos en el eje Y
#   Label en la leyenda
````

# frame.plot()

```python
frame.plot(x,y,label=<Series>,ax,style,kind,xticks,yticks,title,subplots)
```

- (solo para Series) label: referencia para la leyenda
- ax: subfigura en la que dibujar los datos
- style: estilo de la línea
- kind: tipo de gráfico ('bar’, ‘pie’, ‘hist’, ‘area’ ‘line’, ‘barh’, ‘density’, ‘kde’).
- xticks y yticks: valores en los ejes X e Y
- title: string como título
- (solo para DataFrame) subplots: bool si se desean subfiguras separadas para cada columna.
- (solo para DataFrame) 'x' e 'y' se pueden utilizar para seleccionar columnas para cada eje 

[Lista completa](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.html#pandas.DataFrame.plot) de argumentos


````
rand_matrix = np.random.randint(100,size=(10,3))
frame = pd.DataFrame(rand_matrix , columns=list('ABC'))
frame.plot(style=['r-','g-','b-'],xticks=range(-2,12),title='Tipo de lineas',subplots=True)
plt.show()
````

````
# Gráfico de barras
frame.plot(kind='bar',xticks=range(-2,12),title='ipo Barras')
plt.show()
````

````
# Gráfico de área
frame.plot(kind='area',xticks=range(-2,12),title='Tipo área')
plt.show()
````

````
# Gráfico de Pie
df = pd.DataFrame({'mass': [0.330, 4.87 , 5.97], \
                   'radius': [2439.7, 6051.8, 6378.1]}, \
                  index=['A', 'B', 'C'])
df.plot(y='radius',kind='pie',title='Tipo Pie')
plt.show()
````
